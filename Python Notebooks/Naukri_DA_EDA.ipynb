## Automate the above Visualisation

```python
def performance_analysis(data,feature,col_1,col_2,chart):
    '''
    This will return us plots depending upon whatever chart we want
    
    Parameters
    -------
    
    data:dataframe 
    
    feature:column_name that we have to consider for analysis
    
    col1:1st column name that u want to assign to dataframe
    
    col2:2nd column name that u want to assign to dataframe
    
    @chart:if value of chart is 1 ,it represents bar chart
    @chart:if value of chart is 2 ,it represents pie chart

    '''

    dataframe=data[feature].str.strip().value_counts().reset_index()
    dataframe.columns=[col_1,col_2]
    print(dataframe.head(15))

    if chart == 1:
        print("\n This is a barchart \n")
        plt.figure(figsize=(14,6))
        plt.bar(dataframe[col_1][:15], dataframe[col_2][:15])
        plt.xticks(rotation="vertical")
        plt.show()

    elif chart == 2:
        print("\n this is a PieChart \n")
        plt.figure(figsize=(14,6))
        plt.pie(labels=dataframe[col_1][:15],x=dataframe[col_2][:15],autopct="%1.1f%%")
        plt.xticks(rotation="vertical")
        plt.show()
```

```python
print(performance_analysis.__doc__)
```

### median of a max_pay Industry-wise
    1..Using groupby
    2..aggregations is all about (like median, mean, min, max, etc.) on multiple columns at the same time:

```python
data.columns
```

```python
data[["max_pay","industry"]].groupby(["industry"]).median().head(10)
```

```python
data[['max_pay','industry']].groupby(["industry"]).agg({'max_pay':'median'})
```

```python
data[['max_pay','industry']].groupby(["industry"]).median().sort_values(by='max_pay',ascending=False).head(10)
```

```python
data[['max_pay','industry']].groupby(["industry"]).median().sort_values(by='max_pay',ascending=False).head(10).plot(kind='bar',color='green')
plt.title("Top 10 High Paying Industry")
plt.show()
```

```python
data[['min_pay','industry']].groupby(["industry"]).median().sort_values(by='min_pay',ascending=False).head(10).plot(kind='bar',color='pink')
plt.title("Lowest Paying Industries")
plt.show()
```
